### csv to json 

```python
# Code modified from http://www.andymboyle.com/2011/11/02/quick-csv-to-json-parser-in-python/
def csv_to_json(filename):
    '''
    Convert csv file to json file
    Inputs:
        filename(string)
    '''
    # Open the CSV  
    f = open(filename)  
    fieldnames = next(csv.reader(f))
    # Change each fieldname to the appropriate key names
    reader = csv.DictReader( f, fieldnames = fieldnames)  
    # Parse the CSV into JSON  
    out = json.dumps( [ row for row in reader ] )  
    print ("JSON parsed!")  
    # Save the JSON
    f = open(filename.split('.')[0]+'.json', 'w')  
    f.write(out)  
    print ("JSON saved!")
```
